**NETFLIX USER ANALYTICS ASSIGNMENT**

```python
#Import Libraries
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

**PART A: DATASET UNDERSTANDING**

```python
#Q1.Load the dataset and display the first five records
df = pd.read_csv('Dataset 2.csv')
print("First Five Records:")
print(df.head())
```

```python
#Q2.Determine number of rows and columns
print("Number of Rows and Columns:")
print(df.shape)
```

```python
#Q3.Display all column names
print("Column Names:")
print(df.columns)
print("Data Types:")
print(df.dtypes)
print("Summary Statistics:")
print(df.describe())
```

```python
#Q4.Identify numerical and catergorical features
print("Numerical Features:")
print(df.select_dtypes(include=np.number).columns)
print("Categorical Features:")
print(df.select_dtypes(include='object').columns)
```

```python
#Q5.Check missing values
print("Missing Values:")
print(df.isnull().sum())
```

**PART B: EXPLORATORY DATA ANALYSIS**

```python
#Q6.Average age
print("Average Age:")
print(df['Age'].mean())
```

```python
#Q7.Average watch hours per week
print("Average Watch Hours per Week:")
print(df['WatchHoursPerWeek'].mean())
```

```python
#Q8.Average monthly spending
print("Average Monthly Spending:")
print(df['MonthlySpend'].mean())
```

```python
#Q9.Count users in each subscription category
print("Subscription Category Counts:")
print(df['SubscriptionType'].value_counts())
```

```python
#Q10.Percentage of users who renewed
renewal_percentage = (df['SubscriptionRenewed'].value_counts(normalize=True) * 100)
print("Percentage of Users Who Renewed:")
print(renewal_percentage)
```

**PART C: DATA PREPARATION**

```python
#Q11.Convert categorical features to numerical form
le = LabelEncoder()
categorical_columns = [
    'Gender',
    'SubscriptionRenewed',
    'FavoriteGenre',
    'SubscriptionType'
    ]
for column in categorical_columns:
    df[column] = le.fit_transform(df[column])
print("Encoded Dataset:")
print(df.head())
```

```python
#Q12.Define X and y
X = df.drop(columns=['UserID','SubscriptionRenewed'],axis=1)
print("Features:")
print(X.head())
y = df['SubscriptionRenewed']
print("Target Variable:")
print(y.head())
```

```python
#Q13.Split Dataset
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
    )
print("Training Set Shape:")
print(X_train.shape)
print("Testing Set Shape:")
print(X_test.shape)
```

**PART D: DECISION TREE CLASSIFICATION**

```python
#Q14.Train Decision Tree to predict whether a user will renew their subscription
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
print("Decision Tree Model Trained Successfully")
```

```python
#Q15.Accuracy
dt_predictions = dt_model.predict(X_test)
dt_accuracy = accuracy_score(y_test, dt_predictions)
print("Decision Tree Accuracy:")
print(dt_accuracy)
```

```python
#Q16.Confusion Matrix
dt_confusion_matrix = confusion_matrix(y_test, dt_predictions)
print("Decision Tree Confusion Matrix:")
print(dt_confusion_matrix)
```

**PART E: KNN CLASSIFICATION**

```python
#Q17.Train KNN with K=5
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train_scaled, y_train)
knn_predictions = knn_model.predict(X_test_scaled)
knn_accuracy = accuracy_score(y_test, knn_predictions)
print("KNN Accuracy:")
print(knn_accuracy)
print("KNN Model Trained Successfully.")
```

```python
#Q18.Compare KNN and Decision Tree
knn_accuracy = accuracy_score(y_test, knn_predictions)
dt_accuracy = accuracy_score(y_test, dt_predictions)
print("KNN Accuracy:", knn_accuracy)
print("Decision Tree Accuracy:", dt_accuracy)
if knn_accuracy > dt_accuracy:
    print("KNN performs better.")
elif knn_accuracy < dt_accuracy:
    print("Decision Tree performs better.")
else:
    print("Both models have the same accuracy.")
```

**PART F: LINEAR REGRESSION**

```python
#Q19.Train Linear Regression Model to predict monthly spending
df_regression = pd.read_csv('Dataset 2.csv')
for column in categorical_columns:
    df_regression[column] = le.fit_transform(df_regression[column])
X_regression = df_regression.drop(columns=['UserID','MonthlySpend'],axis=1)
y_regression = df_regression['MonthlySpend']
X_train_r, X_test_r, y_train_r, y_test_r = train_test_split(
    X_regression,
    y_regression,
    test_size=0.20,
    random_state=42
    )
lr_model = LinearRegression()
lr_model.fit(X_train_r, y_train_r)
print("Linear Regression Model Trained Successfully")
```

```python
#Q20.Predict monthly spending for a new user and interpret the result
# Example New User:
# Age=25
# Gender=Male
# SubscriptionType=VIP
# WatchHoursPerWeek=20
# DevicesUsed=3
# FavoriteGenre=Sci-Fi
# AdClicks=15
# SubscriptionRenewed=Yes
new_user = pd.DataFrame({
    'Age': [25],
    'Gender': [1],
    'SubscriptionType': [2],
    'WatchHoursPerWeek': [20],
    'DevicesUsed': [3],
    'FavoriteGenre': [4],
    'AdClicks': [15],
    'SubscriptionRenewed': [1]
})
predicted_monthly_spending = lr_model.predict(new_user)
print("Predicted Monthly Spending for the New User:")
print(predicted_monthly_spending)
```
